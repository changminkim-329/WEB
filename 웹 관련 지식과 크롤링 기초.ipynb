## urlopen 
- 다운로드 함수는 아니지만 다른 함수에서 urlopen으로 웹에서 수신된 데이터들을 받고, 그 자체로 다른 함수 인자로 많이 쓰인다.
- 처리용

```python
import urllib.request as req

# urllib 예외처리 메소드 
from urllib.error import URLError, HTTPError
```

```python
# 다운로드 경로 및 파일명

path_list = ['/Users/changmin.kim/Desktop/jupyterNotebook/파이썬 웹 개발/crawlingDataset/test1.jpg',
             '/Users/changmin.kim/Desktop/jupyterNotebook/파이썬 웹 개발/crawlingDataset/index.html'
            ]

# 다운로드 리소스
target_url_list = [
    'http://blogfiles.naver.net/20150612_38/zfghjp12370_1434105617178ExRIE_PNG/%C1%D2%C1%D2_3%BA%CE',
    'https://google.com'
    ]
```

![](https://d22rw8pyewxbkl.cloudfront.net/wp-content/uploads/2008/01/statuscode.png)

```python
for i,url in enumerate(target_url_list):
    try:
        # 웹 수신 정보 읽기
        response = req.urlopen(url)
        
        # 수신내용
        contensts = response.read()
        
        
        print('--------------------------------')
        #상태 정보 중간 출력 
        print('Header Info-{}:{}'.format(i,response.info()))
        print('HTTP Statues Code: {}'.format(response.getcode()))
        print()
        print('--------------------------------')
        
        with open(path_list[i],'wb') as c:
            c.write(contensts)
        
    except HTTPError as e:
        print('Download failed')
        print("HttpError code:",e.code) # http 에러 코드 출력(404)
    except URLError as e:
        print('Download failes')
        print("Url Error Reason:",e.reason)
    else:
        print()
        print("Download succeed")
        
        
```

### 브라우저 구조
- 구조: 웹 콘텐츠에 의미를 부여하고 구조를 형성 : HTML
- 표현: 시각적인 디자인과 레이아웃 표현 : CSS
- 행위: 모든 front-end의 브라웢 상화작용을 담당 : JavaSript

**HTML**
- **Hyper Text Markup Language**
    - Hyper Text: 참조가능한(링크) 텍스트
    - Markup Language: 의미 있는 태그로 콘텐츠를 감싸는 프로그래밍 언어(마크업)
- **웹페이지에서 콘텐츠의 구조를 표현**하기 위해 고안된 **텍스트 포맷**
- HTML은 컨텐츠의 서로 다른 부분들을 태그 등으로 씌우거나 감싸서 다른 형식으로 보이게 하는 형태.
- 대부분의 요소는 또 다른 요소(태그를 포함한 영역)를 포함할 수 있음 -> 부모 자식 관계가 성립.
- HTML은 크게 Html에 Head, Body가 포함된 구조로 되어 있음
    - head는 브라우저 타이틀 바에 출력되는 정보라거나, 연결된 CSS나 JavaScript 파일이 어디에 위치해있는지 알려주는 파일 정보 등이 담겨 있어 문서 자체에 대한 정보만 가지고 있어서 일반 사용자에게는 출력되지 않습니다.
    - body는 눈에 보이는 정보를 가짐

#### Rule Set
HTML페이지 안의 특정 요소들을 어떻게 렌더링(Rendering) 할 것인지 브라우저에게 알려주는 **CSS 문장**.
![](http://www.nextree.co.kr/content/images/2016/09/yrkim-140327-selector-04.png)

- **선택자(Selector)**: Rule Set의 영향을 받는 HTML페이지 안의 특정 element들을 선택하여, 선언 블록(Declaration Block)의 내용을 적용
    - 종류
        - **전체 선택자( * )**: HTML페이지 내부의 모든 태그를 선택.(margin이나 padding값을 초기화하는 등 기본값을 정해둘 때 사용)<br>
        ex) * { margin: 0; text-decoration: none; }
        <br><br>
        - **태그 선택자(p)**: 태그명이 p인 특정 태그를 선택.<br>
        ex) p { background: yellowgreen; color: darkgreen; }
        <br><br>
        - **클래스 선택자(.myClass)**: 클래스 속성값이 myClass으로 지정된 요소를 선택.<br>
        ex) .changminLandclass { background: yellowgreen; color: darkgreen; }<br>
        ex) div.class2 { background: darkgreen; color: yellowgreen; }
            - 마침표 앞에 태그를 붙여주면 범위를 그 특정 태그에 한합니다. 결국 HTML 코드의 두번째
        <br><br>
        - **ID 선택자(#myId)**: id 속성값이 myid로 지정된 요소를 선택.<br>
        : 클래스 선택자랑 키워드만 다르고 남어지는 똑같음

```python
def google_image_downloader(url,google_name):
    # 반드시 google 이미지 주소 이어야 한다.
    
    html = requests.get(url).content
    parser = lxml.html.fromstring(html)
    m = parser.cssselect('table td a img[style]')
    image_urls = []
    for i in range(len(m)):
        images = m[i].get('src')
        image_urls.append(images)
    
    dir_path = os.path.join('/Users/changmin.kim/Desktop/jupyterNotebook/dataset/',google_name+'Image')
    print(dir_path)
    if not(os.path.exists(dir_path)):
        os.makedirs(dir_path)
        
    for i in range(len(image_urls)):
        path = os.path.join(dir_path,google_name+str(i+1)+'.png')
        
        response = req.urlopen(image_urls[i])
        with open(path,'wb') as c:
            c.write(response.read())
```

## 요청 헤더 정보

### HTTP Referer(리퍼러)
- 웹 브라우저로 월드 와이드 웹을 서핑할 때, 하이퍼링크를 통해서 각각의 사이트로 방문시 남는 흔적.
- 웹 사이트의 서버 관리자가 사이트 방문객이 어떤 경로로 자신의 사이트에 방문했는지 알아볼 때 유용하게 사용.
- http header에 기록됨.
- 해당 링크가 걸린 아이템이 존재하는 URL.
- 올바른 리퍼러 정보를 보내지 않으면 문제가 발생할 수도 있는데 웹서버는 자신의 페이지중 일부 페이지에 올바른 리퍼러 정보를 송신하지 않는 브라우저에 대해 엑세스를 Forbidden(403)해 버린다.
- POST같은 요청을 보낼 때, 요청이 어느 주소에서 시작되었는지를 나타내는데 이때 보낸 주소와 받는 주소가 다르면 CORS 문제가 발생
    - ex)Origin : http://goddaehee.tistory.com
    - ex)Referer: http://goddaehee.tistory.com
    
**CORS 문제**: 외부 서버의 API에서 어떤 데이터를 가지고 오려고 할 때 헤더 정보제공 서버에서 접근을 허락하는 헤더 내용이 없으면 발생.
### User-Agent(UA)
- 사용자를 대신하여 일을 수행하는 소프트웨어 에이전트.
- UA만 알아도 어떤 기기/브라우저를 사용하는지 알 수 있음.
- 현재 사용자가 어떤 클라이언트(운영체제, 앱, 브라우저 등)를 통해 요청을 보냈는지 알 수 있음.

### Cookie
- 웹서버가 클라이언트에 쿠키를 저장해 놓았다면 해당 쿠키의 정보를 이름-값 쌍으로 웹서버에 전송한다.

### Accept
- 클라이언트가 허용할 수 있는 파일 형식
- ex)
    - Accept: text/html (HTML 형식인 응답을 처리하겠다)
    - ex2) Accept: image/png, image/gif
    - ex3) Accept: text/*


## 세션(서버와 클라이언트의 인터페이스)

### 개요
웹은 대다수가 HTTP기반으로 동작합니다. 하지만 HTTP가 구현된 방식에서 웹 서버와 클라이언트는 지속적으로 **연결을 유지한 상태가 아니라 요청(request)-응답(response)의 반복**일 뿐이기 때문에, 이전 요청과 새로운 요청이 같은 사용자(같은 브라우저)에서 이루어졌는지를 확인하는 방법이 필요합니다. 이 때 등장하는 것이 ‘쿠키’와 ‘세션’입니다.

## 세션(Session)이란
일정 시간동안 같은 사용자(정확하게 브라우저를 말한다)로 부터 들어오는 일련의 요구를 하나의 상태로 보고 그 상태를 일정하게 유지시키는 기술이다. 방문자의 요청에 따른 정보를 방문자 메모리에 저장하는 것이 아닌 웹 서버가 세션 아이디 파일을 만들어 서비스가 돌아가고 있는 서버에 저장하는 것이다. 서버에 저장되기 때문에 사용자 정보가 노출되지 않는다.

 
## 쿠키(Cookie)란 
- 인터넷 사용자가 어떠한 웹 사이트를 방문할 경우 그 사이트가 사용하고 있는 서버를 통해 인터넷 사용자의 컴퓨터에 설치되는 작은 기록 정보 파일이다.
- Key-Value형식으로 로컬 브라우저에 저장.

## 세션의 동작 방식
1. 클라이언트가 서버에 접속 시 세션 ID를 발급받음.
2. 클라이언트는 세션 ID에 대해 쿠키를 사용해서 저장하고 가짐.
3. 클라리언트는 서버에 요청할 때, 이 쿠키의 세션 ID를 서버에 전달해서 사용.
4. 서버는 세션 ID를 전달 받아서 별다른 작업없이 세션 ID로 세션에 있는 클라언트 정보를 가져온다.
5. 클라이언트 정보를 가지고 서버 요청을 처리하여 클라이언트에게 응답.

## Node란?
HTML DOM은 노드(node)라고 불리는 계층적 단위에 정보를 저장.
![](http://tcpschool.com/lectures/img_js_htmldom.png)

### node의 종류
|노드|설명|
|:---:|:---:|
|**문서 노드(document node**|HTML 문서 전체를 나타내는 노드.|
|**요소 노드(element node)**|모든 HTML 요소는 요소 노드이며, 속성 노드를 가질 수 있는 유일한 노드임.
|**속성 노드(attribute node)**|모든 HTML 요소의 속성은 속성 노드이며, 요소 노드에 관한 정보를 가지고 있음.<br>하지만 해당 요소 노드의 자식 노드(child node)에는 포함되지 않음.|
|**텍스트 노드(text node)**|HTML 문서의 모든 텍스트는 텍스트 노드임.|
|**주석노드(comment node)**|HTML 문서의 모든 주석은 주석 노드임.|

### node간의 관계
![](http://tcpschool.com/lectures/img_js_node_relationship.png)
- 노드 트리의 가장 상위에는 단 하나의 루트 노드(root node)가 존재.
- 루트 노드를 제외한 모든 노드는 단 하나의 부모 노드(parent node)만을 가진다.
- 모든 요소 노드는 자식 노드(child node)를 가질 수 있다.
- 형제 노드(sibling node)란 같은 부모 노드를 가지는 모든 노드
- 조상 노드(ancestor node)란 부모 노드를 포함해 계층적으로 현재 노드보다 상위에 존재하는 모든 노드
- 자손 노드(descendant node)란 자식 노드를 포함해 계층적으로 현재 노드보다 하위에 존재하는 모든 노드

## chrome html로 이미지 저장

```python
def chrome_big_image_parser(driver, image_url, time_sleep, image_name, length=None):
    import time
    from base64 import b64decode
    import urllib.request as req
    import os
    """
    반드시 driver를 실행한 다음에 사용하세요.
    가능하나 length의 길이를 너무 크게 하지 마세요. 구글 ip 정지 먹어요
    """
    
    dir_path = os.path.join('/Users/changmin.kim/Desktop/jupyterNotebook/big_image_data/',image_name+'Image')
    if not(os.path.exists(dir_path)):
        os.makedirs(dir_path)
    try:
        driver.get(image_url)
    except Exception as e:
        driver = webdriver.Chrome('/Users/changmin.kim/Desktop/jupyterNotebook/chrome_driver/chromedriver')
        driver.implicitly_wait(3)
        driver.get(image_url)
    
    g = driver.find_elements_by_xpath('//*[@id="islrg"]/div[1]//div/a[1]/div[1]/img')
    if(length==None):
        length=len(g)
    print(length)
    for i in range(length):
        if(g[i].get_attribute('src')==None):
            time.sleep(time_sleep)
            driver.execute_script('window.scrollBy(0,4000)')
            g = driver.find_elements_by_xpath('//*[@id="islrg"]/div[1]//div/a[1]/div[1]/img')
            print("페이지를 스크롤 합니다.")
            print("다음 페이지에서 이미지의 수는: "+len(g))
            continue
            
        scheme = g[i].get_attribute('src').split('/')[0]
        if(scheme=='data:image'):
            print(i,"번째 사진 download",g[i].get_attribute('jsname'))
            header, encoded = g[i].get_attribute('src').split(",", 1)
            data = b64decode(encoded)
            with open(dir_path+"/"+image_name+str(i)+'.png', "wb") as f:
                f.write(data)
            
        if(scheme=='https:'):
            url = req.urlopen(g[i].get_attribute('src'))
            print(url)
            url_content = url.read()
            with open(dir_path+"/"+image_name+str(i)+'.png', "wb") as f:
                f.write(url_content)
        time.sleep(time_sleep)
```

## URI란?
**개요**<br>
클라이언트가 서버에게 자원을 요청할 때 자원의 위치는 일반적으로 URL정적)을 사용해서 나타낸다. <br>
하지만, **검색엔진을 검색한 결과와 같이 동적으로 생성된 정보는 URL로 표현할 수 없다.**
- Uniform Resource Identifier
- 동적문서를 요청하는 경우, **URL**과 **사용자 질의(query)**를 합친 것.
- **URL과 URI의 차이점**
    - **URL**은 자원의 위치만 나타낸다.
    - **URI**는 자원의 위치에 전달될 인자까지 나타낸다.
## 웹문서 종류
- 정적(Static)문서.
- 동적(Dynamic)문서.
- 엑티브(Active)문서.

**정적문서**<br>
- 서버에서 생성된 고정문서.
- 클라이언트는 문서의 복사본만을 가짐.
- 파일이 생성될때 내용이 결정되므로 클라이언트는 일반적으로 수정이 불가능.
- **HTML**
    - 웹페이지 언어.
    - 링크를 지원하여 새로운 페이지로 이동할 수 있는 기능 제공.
    - 각종 태그 효과.
    
**동적문서**<br>
- 브라우저가 문서를 요청할 때마다 웹 서버에 의해 생성.
    - 요청시, 동적문서를 생성시키는 **응용프로그램이나 스크립트** 실행.
    - 이후 요청을 한 브라우저에게 응답
    - 요청에 의해 생성되므로, 요청에 따라 내용이 달라질 수 있음
- 클라이언트가 아닌 서버에서 실행해서 출력.
- 스크립트를 이용한 동적문서 생성 예.
    - **PHP**(Perl Hypertext Preprocessor)
    - **JSP**(Java Server Pages)
    - **ASP**(Active Server Pasges)
    - **ColdFusion

- **CGI**(공통 게이트웨이 인터페이스)
    - Comman Gateway Interface: 규칙, 엑세스, 용어및 변수, 함수의 집합.
    - 동적문서를 생성하고 처리하는 기술(프로그래밍 언어가 아님)
        - 단지, 사용자가 C,C++등의 언어 중 하나를 사용할 수 있도록 함.
    - 입력(input): URL에서 '?'앞부분은 수행될 프로그램 뒷부분은 입력으로 해석
    - 출력(output): 출력되는 형태는 HTML 구조를 갖는 문서, 상태코드, 이진 데이터 등.

**엑티브문서**
- 클라이언트 사이에서 수행되는 문서(화면에서 움직이는 그림등)
- 사용자와 상호작용하는 프로그램.
- 서버가 아닌 클라이언트에서 실행.
- 사용절차:
    - 1. 브라우저가 엑티브 문서를 요청
    - 2. 서버는 문서의 복사본이나 스크립트를 전송.
    - 3. 문서는 클라이언트 사이트에서 수행.
- **Java Script**
    - 컴파일 과정없이 실행 가능.
    - 동적문서의 스크립트 개념이 액티브 문서에도 사용 가능.
    - 스크립트 -> 소스코드인 바이너리 형태.
    - 자바 스크립트-> 자바와 닮은 점이 있는 스크립트 언어.
    
- **자바 애플릿**
    - 서버에서 자바로 작성된 프로그램.
    - 컴파일된 바이너리 형태.
    - 브라우저는 이 애플릿의 인스턴스를 생성하고 이를 수행.


## Data URIs란?
- **Data URI scheme**의 약자로, `data:` 스킴이 접두어로 붙은 URI.
- 작은 파일을 문서 내에 인라인으로 임베디드('내장:')하기 위한 URI 스킴이다. 
- **문법**<br>
- 주로 아이콘 등의 작은 이미지 파일에 사용.
`data:[<mediatype>][;base64],<data>`
- `mediatype`: MIME 타입. 생략되었을때, 기본 값 text/plain;charset=US-ASCII
- `base64`: 데이터가 텍스트가 아닌 경우 base64 인코딩된 데이터가 필요하다.

- 특징
    - 소스코드 내 삽입
        - 추가적인 HTTP요청이 일어나지 않는다.
    - 용량
        - base64 포맷으로 인코딩하면 원본 파일보다 더 용량이 커진다.
    - 긴 문자열
        - 브라우저에 따라 문자 길이에 제한이 있기 때문에, 용량이 큰 이미지 등은 처리하지 못할 수 있다.
        - 소스코드 상에 삽입하면 소스 가독성이 떨어질 수 있다.
        
- **Sample(URL에 쳐보기)**<br>
    - data:,Hello%2C%20World!
    - data:text/plain;base64,SGVsbG8sIFdvcmxkIQ%3D%3D

## XML과 HTML의 차이점

- #### HTML이 데이터의 표현에 그 목적을 두고 있다면 XML은 데이터 교환을 위한 구조정의에 그 목적을 두고 있다.

### XML
- eXtensible Markup Language
- 데이터 저장과 전송을 목적으로 디자인된 만들어진 마크업 언어. == json
- HTML 의 태그는 각각 정해진 의미를 가지지만, XML은 사용자가 태그를 정의해서 사용한다.(사전 정의 태그가 존재 X)
- HTML이 인터넷 웹 환경에서 작동되는 언어라고 본다면, XML은 어느 특정 환경에 구애를 받지 않음.
- XML을 다른 환경에서 사용하기 위해 객체 단위로 바꾸어서 사용.
    - ex) 안드로이드 XML
    - json: Javascript Object notation
    - <태그 이름> -> 객체 이름
- HTML이 데이터와 그 표현을 동시에 하고 있다면, XML은 데이터만을 가지고 있을 뿐 그 표현은 하지 않음.

`<TBL_StockInfo>
     JongName="NAVER보통주"
     CurJuka="113,000"
     StockDungRak="5"
 </TBL_StockInfo>`
 
 - 위의 xml를 객체로 바꾼다면 TBL_StockInfo 클래스 이름이 되고, JongName,CurJuka,..등이 클래스의 속성.

```python
next_chapter_page = 0 
present_page_state = 0
previous_page_state = 0
page_number = 0
most_review_word = {}

while True:
    present_page_state = driver.find_elements_by_css_selector('#btfTab > ul.tab-contents > li.product-review > div > div.sdp-review__article.js_reviewArticleContainer > section.js_reviewArticleListContainer > div.sdp-review__article__page.js_reviewArticlePagingContainer')
    present_page_state = present_page_state[0].text.split()
    page_number = len(present_page_state)
    present_page_state = present_page_state[0]
    if(present_page_state!=previous_page_state):
        print("페이지",present_page_state)
        for i in range(page_number):
            pageMachine = driver.find_elements_by_css_selector("div.sdp-review__article__page.js_reviewArticlePagingContainer")
            pageMachine = pageMachine[0].find_elements_by_css_selector("button.sdp-review__article__page__num")
            g = pageMachine[i]
            g.click()
            time.sleep(1)
            
            review_data = driver.find_elements_by_css_selector(content+list_review)
            for i in range(len(review_data)):
                review_content = review_data[i].text.split()
                for i in review_content:
                    try:
                        most_review_word[i] += 1

                    except:
                        most_review_word[i] = 1
        next_chapter_page = driver.find_element_by_css_selector('#btfTab > ul.tab-contents > li.product-review > div > div.sdp-review__article.js_reviewArticleContainer > section.js_reviewArticleListContainer > div.sdp-review__article__page.js_reviewArticlePagingContainer > button.sdp-review__article__page__next.js_reviewArticlePageNextBtn')
        next_chapter_page.click()
        time.sleep(1)
        previous_page_state = present_page_state
        
    else:
        print('finish')
        break
```
