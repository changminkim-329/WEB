## CSS
- Cascading Style Sheet
- 디자인 영역
- **형식**: 태그, 클래스 및 id등{스타일 속성:속성값};
    - ex) div.person{color:green,font-size:medium};

**CSS Media Quary**
- 미디어 쿼리란?
: 각 디바이스들은 각기 다른 해상도와 화면 크기를 가지고 있어 각각의 화면에 맞게 출력을 할 필요성이 생겼는데 이것을 도와주는 역할을 하는 것.
- 반응형 웹을 구현 가능하게 함.
- **형식**: @media [only 또는 not] all and (조건문) {실행문}
    - @media: 미디어 쿼리가 시작됨을 선언
    - [only 또는 not]: 생략 했을 시 only로 초기화 되며, only는 뒤에 오는 조건을 성립하는 경우의 의미고, not은 그 반대다.
    - all: 미디어 쿼리를 해석해야 할 대상 미디어를 선언한 것이며, all 이면 모든 미디어가 이 구문을 해석해야 한다. all 키워드 대신 screen 또는 print와 같은 특정 미디어를 구체적으로 언급할 수도 있다.
        - ex) @media all and (max-width:767px){...}; 사용자 해상도가 767px이하일때 성립.

## DOM
- **Document Object Tree Model**
    - 넓은 의미: 웹 브라우저가 HTML 페이지를 인식하는 방식.
    - 좁은 의미: html문서의 태그들을 JavaScript가 이용할 수 있는 객체(object)들로 만들어 문서 객체 집합으로 만든 것.
- HTML, XML 문서의 프로그래밍 interface.
- html 실행시 태그의 관계에 맞게 객체로 바뀐다.(객체트리[속성/메소드] in Memory)
- DOM 은 웹 페이지의 객체 지향 표현이며, 자바스크립트와 같은 스크립팅 언어를 이용해 DOM 을 동적으로 수정할 수 있다.

#### 시텐틱(Semantic) 태그
- **웹페이지의 구조와 정보의 의미를 담고 있는 태그.**
- HTML5에 도입된 태그.
- 웹페이지 구성을 위한 공간분할태그.
- 태그만으로 문서를 **쉽게 이해할 수 있도록 의미를 부여**한 태그
- 검색엔진이 더욱 논리적으로 해당컨텐츠를 읽을 수 있도록 문서구조를 논리적으로 표현.

___
`<header>`
- **의미: 머리말 영역**
- 페이지(혹의 내용)상단의 머리말을 지정 
- 페지지 제목, 소개 글이나 로고 이미지 등이 포함.
- body, section, article 태그 안에도 여러 번 포함 가능.
___
`<nav>`
- **의미: 메뉴 영역**
- 페이지 이동을 위한 메뉴영역을 지정.
- 이전페이지, 다음페이지, 특정페이지나 사이트에 대한 연결등 네비게이션 요소로 구성.
- header, aside footer 태그 안에도 여러 번 포함 가능.
___
`<aside>`
- **의미: 보조 영역**
- 페이지의 좌,우측면 공간 같은 보조영역을 지정.
- 광고나 즐겨찾기링크,관련이미지 정보 등을 제공.
- section, article태그안의 내용과 간접적으로 관련된 내용을 포함.
___
`<section>`
- **의미: 형식적 구분 영역**
- 기능보단 의미(구부)에 중점을 둔 태그.
- 하나의 제목(`<h1>~<h6>태그`)을중심으로 내용들을 그룹화(section별 제목).
- section태그안에 또 다른 section 태그를 포함한 다양한 태그들이 중첩되어 사용.
___
`<article>`
- **의미: 내용적 구분 영역**
- 독립적인 개별 내용 영역을 지정.
- 블로그나 댓글, 신문,잡지의 기사등을 제공.
- 별도로 배포되거나 재사용 가능한 내용으로 그룹화.
- `<section>태그와 <article>태그 사이의중첩가능`
___
`<footer>`
- **의미: 꼬리말 영역**
- 페이지(혹은내용) 하단의 꼬리말을 지정.
- 작성자와 작성날짜, 저작권등 웹페이지 관련 추가 정보들을 포함.

## FORM(폼)
- 사용자가 웹 사이트로 정보를 보낼 수 있는 요소들은 모두 폼(form) -> get, post 방식
- 폼과관련된대부분의작업들은정보를저장하거나검색하거나수정하는일들
- 아이디를입력하는텍스트필드나버튼같은폼의형태를만드는것은HTML 태그
- 폼에입력한사용자정보를처리하는것은서버프로그램에서처리.

`<form [속성="속성값"]> 여러 폼 요소 </form>`
#### 속성
- method: 사용자가 입력한 내용들을 서버 쪽 프로그램으로 어떻게 넘겨줄지 지정.(get, post:대부분 post 사용)
- action: 서버상의 프로그램을 지정
- target: action태그에서 지정한 스크립트 파일을 현재 창이 아닌 다른 위치에서 열도록 지정.
- name: 폼의 이름 지정

## 미디어 쿼리(Media Query)
- CSS3 모듈 중 하나로, 사이트에 접속하는 기기 별로 적용할 CSS 파일을 다르게 지정해 주는 기술.
- 미디어의 화면의 크기와 환경을 감지하여(쿼리) 웹사이트를 변경하는 기술.
- `@media [ONLY|NOT]미디어 유형[and|,](조건문){실행문}`(대소문자 구분x)

- 테이블의 형식대로 row 클래스와 각종 col-* 클래스들로 자유롭게 구성이 가능
- 자유롭게 구성한 페이지들은 Bootstrap에서 알아서 크기에 따라 반응형으로 동작하도록 만드는 것
- row 클래스는 container나 container-fluid 안에 있어야 정상적인 배열이나 패팅을 지원해준다.

- **GridSystem의 규칙**
    - row 클래스는 가로로 그룹 지을 칼럼들의 집합이다.
    - 내용은 col-* 클래스 안에 있어야 하며, row의 직속 자식 요소로 배치해야 한다.

    - 칼럼은 총 12칼럼이 있는 것으로 정의하여 각 배치할 %에 따라서 클래스를 결정하면 된다. 예를 들어, 같은 넓이의 3개의 칼럼을 쓰고자 한다면 3개의 .col-xs-4 칼럼을 쓰면 된다.

    - 12칼럼이 넘어가면 새로운 줄로 칼럼이 배치된다.
